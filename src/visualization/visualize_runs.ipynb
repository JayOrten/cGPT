```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Load the data
df = pd.read_csv('/home/jo288/nobackup/autodelete/cGPT/runs/109/csv_logs/version_0/metrics.csv')
df2 = pd.read_csv('/home/jo288/nobackup/autodelete/cGPT/runs/110/csv_logs/version_3/metrics.csv')
```

```python
df2.head()
```

```python
# Plot val_loss by removing rows with NaN and plotting with step column
df = df.dropna(subset=['val_loss'])
df = df.reset_index(drop=True)
df['val_loss'] = df['val_loss'].astype(float)

df2 = df2.dropna(subset=['val_loss'])
df2 = df2.reset_index(drop=True)
df2['val_loss'] = df2['val_loss'].astype(float)

# Plot the data
sns.set(style='whitegrid')
plt.figure(figsize=(12, 6))
plt.plot(df['step'], df['val_loss'], label='457M')
plt.plot(df2['step'], df2['val_loss'], label='2B')

plt.xlabel('Step')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
