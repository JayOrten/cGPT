```python
import chess.pgn
from pathlib import Path
```

```python
files = Path('../../dataset/raw/lichess-games/').glob('*')
```

```python
for file in files:
    print(file)
    name = str(file).split('\\')[5].split('.')[0]
    print(name)
    with open(file) as pgn:
        game = chess.pgn.read_game(pgn)
        while game is not None:
            moves = []
            board = game.board()

            for move in game.mainline_moves():
                moves.append(board.san(move))
                board.push(move)
            
            # Add game moves as line to file
            new_file_name = '../../dataset/raw/processed_games/' + name + '.txt'
            with open(new_file_name, 'a') as f:
                f.write(' '.join(moves))
                f.write('\n')

            game = chess.pgn.read_game(pgn)
```

```python
# Iterate through all txt files in the processed games folder and add each line as a row in a single csv files for games
import csv
import os

with open('../../dataset/raw/processed_games.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    for file in os.listdir('../../dataset/raw/processed_games/'):
        with open('../../dataset/raw/processed_games/' + file) as f:
            for line in f:
                writer.writerow([line.strip()])


```
